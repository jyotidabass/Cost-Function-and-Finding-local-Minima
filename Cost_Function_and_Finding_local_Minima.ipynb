# **Cost Function Example**

In this example, we define a simple Cost Function cost_function(x) = (x - 5)**2 + 3, which has a minimum at x=5. We generate 100 values of x between 0 and 10, calculate the corresponding Cost Function values, and plot the result.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define a simple Cost Function
def cost_function(x):
  return (x - 5)**2 + 3  # This is a quadratic function with a minimum at x=5

# Generate dummy data
x_values = np.linspace(0, 10, 100)  # 100 values between 0 and 10

# Calculate the Cost Function values
cost_values = [cost_function(x) for x in x_values]

# Plot the Cost Function
plt.plot(x_values, cost_values)
plt.xlabel('x')
plt.ylabel('Cost')
plt.title('Cost Function')
plt.show()
```

# **Finding Local Minima Example**

In this example, we define a Cost Function with multiple local minima cost_function(x) = np.sin(x) + 0.5 * (x - 3)**2. We define an initial guess x0=2 and use the minimize function from SciPy to find the local minimum using gradient descent. We plot the Cost Function with the local minimum marked.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define a simple Cost Function with multiple local minima
def cost_function(x):
  return np.sin(x) + 0.5 * (x - 3)**2  # This function has multiple local minima

# Define the initial guess for the optimization
x0 = 2

# Minimize the Cost Function using gradient descent
res = minimize(cost_function, x0, method='SLSQP')

# Print the result
print(f"Local Minimum: x = {res.x[0]}, Cost = {res.fun}")

# Plot the Cost Function with the local minimum
x_values = np.linspace(-5, 5, 100)
cost_values = [cost_function(x) for x in x_values]
plt.plot(x_values, cost_values)
plt.plot(res.x[0], res.fun, 'ro')  # Plot the local minimum
plt.xlabel('x')
plt.ylabel('Cost')
plt.title('Finding Local Minima')
plt.show()
```
